```python
from ultralytics import YOLO
import cv2 as cv
import os
```

```python
# создем модель с готовыми весами
model = YOLO("yolo11s.pt")
```

```python
# путь к видео на вход
path_in = "/kaggle/input/birdie/video.mp4"
# путь к видео на выход
path_out = "birdie_out.mp4"
```

```python
# для захвата видео
cap = cv.VideoCapture(path_in)

# FPS
fps = int(cap.get(cv.CAP_PROP_FPS))
# ширина
width = int(cap.get(cv.CAP_PROP_FRAME_WIDTH))
# высота
height = int(cap.get(cv.CAP_PROP_FRAME_HEIGHT))
# кодек
fourcc = cv.VideoWriter_fourcc(*'mp4v')

# для записи результирующего видео
out = cv.VideoWriter(path_out, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    
    # проверяем, что картинка доступна
    if not ret:
        break

    # детектим объекты
    results = model(
        frame,  # текущий кадр
        verbose=False   # прячем неимоверно длинный отчет о результатах
    )

    # аннотируем
    annotated_frame = results[0].plot()  # получаем кадр с bbox

    # Запись кадра в выходное видео
    out.write(annotated_frame)

# Закрытие ресурсов
cap.release()
out.release()
```
