```python
from ultralytics import YOLO
import cv2 as cv
import os
```

```python
# —Å–æ–∑–¥–µ–º –º–æ–¥–µ–ª—å —Å –≥–æ—Ç–æ–≤—ã–º–∏ –≤–µ—Å–∞–º–∏
model = YOLO("yolo11s.pt")
```

```python
# –ø—É—Ç—å –∫ –≤–∏–¥–µ–æ –Ω–∞ –≤—Ö–æ–¥
path_in = "/kaggle/input/birdie/video.mp4"
# –ø—É—Ç—å –∫ –≤–∏–¥–µ–æ –Ω–∞ –≤—ã—Ö–æ–¥
path_out = "birdie_out.mp4"
```

```python
# –¥–ª—è –∑–∞—Ö–≤–∞—Ç–∞ –≤–∏–¥–µ–æ
cap = cv.VideoCapture(path_in)

# FPS
fps = int(cap.get(cv.CAP_PROP_FPS))
# —à–∏—Ä–∏–Ω–∞
width = int(cap.get(cv.CAP_PROP_FRAME_WIDTH))
# –≤—ã—Å–æ—Ç–∞
height = int(cap.get(cv.CAP_PROP_FRAME_HEIGHT))
# –∫–æ–¥–µ–∫
fourcc = cv.VideoWriter_fourcc(*'mp4v')

# –¥–ª—è –∑–∞–ø–∏—Å–∏ —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–µ–≥–æ –≤–∏–¥–µ–æ
out = cv.VideoWriter(path_out, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    
    # –ø—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ –∫–∞—Ä—Ç–∏–Ω–∫–∞ –¥–æ—Å—Ç—É–ø–Ω–∞
    if not ret:
        break

    # –¥–µ—Ç–µ–∫—Ç–∏–º –æ–±—ä–µ–∫—Ç—ã
    results = model(
        frame,  # —Ç–µ–∫—É—â–∏–π –∫–∞–¥—Ä
        verbose=False   # –ø—Ä—è—á–µ–º –Ω–µ–∏–º–æ–≤–µ—Ä–Ω–æ –¥–ª–∏–Ω–Ω—ã–π –æ—Ç—á–µ—Ç –æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞—Ö
    )

    # –∞–Ω–Ω–æ—Ç–∏—Ä—É–µ–º
    annotated_frame = results[0].plot()  # –ø–æ–ª—É—á–∞–µ–º –∫–∞–¥—Ä —Å bbox

    # –ó–∞–ø–∏—Å—å –∫–∞–¥—Ä–∞ –≤ –≤—ã—Ö–æ–¥–Ω–æ–µ –≤–∏–¥–µ–æ
    out.write(annotated_frame)

# –ó–∞–∫—Ä—ã—Ç–∏–µ —Ä–µ—Å—É—Ä—Å–æ–≤
cap.release()
out.release()
```
